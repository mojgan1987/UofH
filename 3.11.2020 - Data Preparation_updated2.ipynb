# PCA: Pricipal Component Analysis

More mathematical background and a deeper dive into PCA can be found in: <br>
    *Introduction to Statistical Learning* by Gareth James et al. (Section 10.2)

Find the link on Canvas.

* Undupervised statistical technique 
* examines interrelations among a set of variables
* to identify underlying structres
* also known as general factor analysis

* Factor analysis determines orthogonal lines of best fit.
* Orthogonal: prependicular in n-dimensional space (at the right angles or 90 degrees)
* n-dimensional variable sample space: as many dimensions as there are variables    

````python
from IPython.display import Image
from IPython.core.display import HTML 
Image(url= "https://datascienceplus.com/wp-content/uploads/2019/09/Capture2-490x386.png")
````

suppose you have 30 features column in a data frame so it will help to reduce the number of features making a new feature which is the combined effect of all the feature of the data frame.

Components: a linear transformation that chooses a variable system where 
* the greatest variance of the dataset comes to lie on the first axis
* the second greatest variance on the second axis
* and so on
this process will allow us to reduce the number of variables used in an analysis.

**Note!** components are uncorrelated, since in the same space they are orthogonal.

Feature scaling: converting all numerical data into same scale or range. <br>
Take a look at the dataframe and check teh range of teh data under different columns. <br>

Why we need feature scaling?
1. most algorith use Euclidean distance: uncertainlty
2. having smaller differences in range decreases our processing time

Our data distribution will have a mean value 0 and standard deviation of 1.

There are different techniques of feature scaling, 
1. normalization: Normalization is a scaling technique in which values are shifted and rescaled so that they end up ranging between 0 and 1. It is also known as Min-Max scaling.
2. standardization: Standardization is another scaling technique where the values are centered around the mean with a unit standard deviation. This means that the mean of the attribute becomes zero and the resultant distribution has a unit standard deviation.

### PCA
1. instantiate a PCA object 
2. find the principal component using the fit method
3. apply the rotation and dimensionality reduction by calling transfrom

Note: we can specify the number of components we want to keep

````python
from sklearn.decomposition import PCA
````

````python
pca = PCA(n_components=2)
````

````python
pca.fit(scaled_data)
````

````python
# transorm the data to its first principal components
x_pca = pca.transform(scaled_data)
````

````python
scaled_data.shape
````

````python
x_pca.shape
````

````python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0], x_pca[:,1], c=cancer['target'])
plt.xlabel('First Principal Component')
plt.ylabel('First Principal Component')
````

````python
# with only 2 principal components that are driven from 30 columns, we see a very clear sepeartion of the type of tumors
# kind of a compression algorithm 
keep a lot of information and explain a lot of variances
````

````python
# the components do not relate one-one to a specific feature in the data 
components correspond to combinations of the original feature
components are stored as an attribute of the PCA object
````

````python
# arrays of actual components
# each row is a principal component and each column relates back to the original features
pca.components_
````

````python
# visualize with a heatmap the relastionship between the correalation of various features and PCA
df_comp = pd.DataFrame(pca.components_, columns=cancer['feature_names'])
````

````python
df_comp
````

````python
plt.figure(figsize=(12,6))
sns.heatmap(df_comp)
````
